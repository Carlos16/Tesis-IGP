#Influence of $k_{RC}$ and $k_{CP}$ over the invasibility functions

#Scenario 2

\begin{equation}
    \frac{\partial \varphi_c}{\partial k_{RC}} = \epsilon_1 ( \frac{\partial K}{\partial k_{RC}} \alpha_1 + K \frac{\partial \alpha_1}{\partial k_{RC}})  -   \frac{\partial q_1}{\partial k_{RC}}
\end{equation}

Where :

\begin{equation}
    \begin{aligned}
        \frac{\partial K}{\partial k_{RC}} &= (1-w) K k_{RC}^{-1}  > 0\\
        \frac{\partial \alpha_1}{\partial k_{RC}} &=  \frac{\alpha_1}{g_1 \Phi} ( \frac{\partial g_1}{\partial k_{RC}} \Phi + g_1 \frac{\partial \Phi}{\partial k_{RC}})
    \end{aligned}
\end{equation}

For the different foraging strategies we have that :

def : 
\begin{equation}
    h_{RC} := \sqrt{1+ (\frac{v_{0,R}}{v_{0,C}}) k_{RC}^{2p_v}\Delta^2_{RC}}
\end{equation}

$h_{RC}$ will change with the thermy of consumer and resource but the qualitative influence of $k_{RC}$ will be the same

\begin{equation}
    \begin{aligned}
    &\frac{\partial g_1}{\partial k_{RC}}
    \begin{cases} 
    ((D_R -1)p_d + \frac{p_v(h_{RC}^2 - 1)}{h_{RC}^2})k_{RC}^{-1} g_1 & ,f = Active \\
    (p_v + (D_R -1)p_d )k_{RC}^{-1}g_1 &,f=Sit\\   
    ( D_R - 1)p_d k_{RC}^{-1}g_1 &, f= Grazing
    \end{cases}
    \\
    &\frac{\partial \Phi}{\partial k_{RC}} = -\frac{b k_{RC}^{b-1}}{(1+k_{RC}^b)^2} = - b \Phi^2 k_{RC}^{b-1}
    \end{aligned}
\end{equation}

In all cases:
\begin{equation}
    \begin{aligned}
    \frac{\partial g_1}{\partial k_{RC}} > 0 \\
    \frac{\partial \Phi}{\partial k_{RC}} <0 
    \end{aligned}
\end{equation}

And for $\alpha_1$ we have that :

\begin{equation}
    \frac{\partial \alpha_1}{\partial k_{RC}} 
    \begin{cases} 
    \frac{k_{RC}^{-1}\alpha_1}{(1+k_{RC}^b)^2}((1+k_{RC}^b)^2((D_R -1)p_d + \frac{p_v(h_{RC}^2 - 1)}{h_{RC}^2}) - b k_{RC}^{b} )& ,f = Active \\
    \frac{k_{RC}^{-1}\alpha_1}{(1+k_{RC}^b)^2}( (1+k_{RC}^b)^2(p_v + (D_R-1)p_d)  - b k_{RC}^b) &,f = Sit\\
    \frac{k_{RC}^{-1}\alpha_1}{(1+k_{RC}^b)^2}( (1+k_{RC}^b)^2(D_R-1)p_d  - b k_{RC}^b) &,f = Grazing 
    \end{cases}
\end{equation}

#Influence over attack rates

## $\alpha_1 - \alpha_2$

\begin{equation}
    \frac{\alpha_1}{\alpha_2} = \frac{\alpha_{1,0}v_{0,C}m_C^{p_v+2(D_R-1)p_d - 1}}{\alpha_{2,0}v_{0,P}m_P^{p_v + 2(D_R-1)p_d -1}} \frac{g_1(k_{RC})\Phi(k_{RC})}{g_2(k_{RP})\Phi(k_{RP})}
\end{equation}

For *grazing,grazing* combination and assuming $\alpha_{1,0} = \alpha_{2,0}$ , $v_{0,C} =v_{0,P}$  we have :

\begin{equation}
    \frac{\alpha_1}{\alpha_2} = \frac{1 + (k_{RC}k_{CP})^b}{ k_{CP}^{1-(p_v + (D_R-1)p_d)} (1+k_{RC}^b)}
\end{equation}

We have the following behavior (note that it equals one when $k_{CP} = 1$)
\begin{equation}
    \begin{aligned}
    \frac{\alpha_1}{\alpha_2} >1 &\iff 1 - k_{CP}^{1 - (p_v + (D_R-1)p_d)} > k_{RC}^b(k_{CP}^{1 - (p_v + (D_R-1)p_d)} - k_{CP}^b) \\ &\iff \frac{1 + k_{RP}^b}{(1+k_{RC}^b)} > k_{CP}^{1 - (p_v + (D_R-1)p_d)}
    \end{aligned}
\end{equation}

$b > 1 - (p_v + (D_R - 1) p_d) $

\begin{equation}
    \frac{\alpha_1}{\alpha_2} > 1 \iff \begin{cases} \frac{ 1 - k_{CP}^{1 - (p_v + (D_R-1)p_d)}}{K_{CP}^{1 - (p_v + (D_R-1)p_d)} - k_{CP}^b} > k_{RC}^b &, k_{CP} <1 \\ \frac{k_{CP}^{1 - (p_v + (D_R-1)p_d)} - 1}{k_{CP}^b - k_{CP}^{1 - (p_v + (D_R-1)p_d)}} < k_{RC}^b &, k_{CP} >1 \end{cases}
\end{equation}

$b < 1 - (p_v + (D_R-1)p_d)$ 

\begin{equation}
    \frac{\alpha_1}{\alpha_2} > 1 \iff \begin{cases} \frac{ 1 - k_{CP}^{1 - (p_v + (D_R-1)p_d)}}{K_{CP}^{1 - (p_v + (D_R-1)p_d)} - k_{CP}^b} > k_{RC}^b &, k_{CP} > 1 \\ \frac{k_{CP}^{1 - (p_v + (D_R-1)p_d)} - 1}{k_{CP}^b - k_{CP}^{1 - (p_v + (D_R-1)p_d)}} < k_{RC}^b &, k_{CP} < 1 \end{cases}
\end{equation}
given that $k_{RC} > 0$ then the above conditions results in :
\begin{equation}
    \frac{\alpha_1}{\alpha_2} > 1 \iff  k_{CP} < 1 
\end{equation}


$b = 1 - (p_v + (D_R-1)p_d)$ 

\begin{equation}
    \frac{\alpha_1}{\alpha_2} >1 \iff 1 - k_{CP}^{1 - (p_v + (D_R-1)p_d)} > 0  \iff k_{CP} > 1
\end{equation}

Let us define : 
\begin{equation} h  = \frac{k_{CP}^{1 - (p_v + (D_R-1)p_d)} - 1}{k_{CP}^b - k_{CP}^{1 - (p_v + (D_R-1)p_d)}}\end{equation}

\begin{equation}
    \frac{\partial h }{\partial k_{CP} } = k_{CP}^{-(p_v + (D_R-1)p_d)} ((1-(p_v+(D_R-1)p_d) (k_{CP}^b - 1) - b k_{CP}^b) + bk_{CP}^{b-1}
\end{equation}

\end{equation}

\begin{equation}
    \frac{\alpha_2}{\alpha_3}  > 1 \iff k_{RC}^{(D_R-1)p_d} > \frac{\sqrt{1+k_{CP}^{2p_v}}}{1 + k_{CP}^b}(1+ k_{RP}^b)
\end{equation}
